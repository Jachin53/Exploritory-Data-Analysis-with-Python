```python
### Conversting the DATE to a data type date
df["Date"] = pd.to_datetime(df["Date"],format="%Y-%m-%d")
display(df)
```

```python
### Extracting Day,Month,Year from the Date Column
df["Day"] = df["Date"].dt.day_name()
df["Month"] = df["Date"].dt.month_name()
df["Year"] = df["Date"].dt.day_name()
display(df)
```

```python
###Check the max age and the min age
df["Age"].max()
df["Age"].min()
```

```python
### Create a case statement for Age-Classification
Conditions = [
    (df["Age"].between(18,25)),   
    (df["Age"].between(26,35)),  
    (df["Age"].between(36,64)),   
    (df["Age"] >= 65)            
]
Values = ["Young-Adult","Adult","Senior","Elderly"]
df["Age_Classification"] = np.select(Conditions, Values, default="Unknown")
print(df)

```

```python
## Case statement for spending buckets
df["Spending_Bucket"] = "Low"
df.loc[df["Total Amount"] >= 100, "Spending_Bucket"] = "Medium"
df.loc[df["Total Amount"] > 200, "Spending_Bucket"] = "High"
print(df)
```

```python
df.head(100)
```

```python
### Retrieve the Total Revenue 
df["Total_Revenue"] = df["Total Amount"] * df["Quantity"]
display(df)
```

```python
### Average Total_Revenue
df["Total_Revenue"].mean()
```

```python
### Total Revenue GROUPED BY Gender
df.groupby(["Gender"], as_index=False).agg(total_revenue=("Total_Revenue", "sum"))
```

```python
### Total_Revenue GROUP BY Age_Classification
df.groupby(["Age_Classification"], as_index=False).agg(total_revenue=("Total_Revenue", "sum")).sort_values("total_revenue",ascending=False)
```

```python
### Total Revenue GROUP BY Product Category
df.groupby(["Product Category"], as_index=False).agg(total_revenue=("Total_Revenue", "sum")).sort_values("total_revenue",ascending=False)

```

```python
### Average Total Revenue GROUP BY Product Category
df.groupby(["Product Category"], as_index=False).agg(Avg_Revenue=("Total_Revenue", "mean")).sort_values("Avg_Revenue",ascending=False)

```
